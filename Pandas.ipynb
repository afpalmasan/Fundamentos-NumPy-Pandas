## MERGE | CONCAT

---



```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
pd.concat([df1,df2], ignore_index=True) # concadena por default en axis 0
```

```python
pd.concat([df1,df2], ignore_index=True, axis=1) # se concatea de forma horizontal
```

## Merge

```python
left = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],})

right = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
right
```

```python
left.merge(right, on='Key') # realiza el merge indicando la columna match o llave
```

```python
left = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],})

right = pd.DataFrame({'Key_2':['k0','k1','k2','k3'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
```

```python
left.merge(right, left_on='Key', right_on='Key_2') # merge de 2 dataframe con 2 llaves distintas
```

```python
left = pd.DataFrame({'Key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],})

right = pd.DataFrame({'Key_2':['k0','k1','k2',np.nan],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']})
left.merge(right, left_on='Key', right_on='Key_2') # al no tener una llave para realizar merge realiz un inner join
```

```python
left.merge(right, left_on='Key', right_on='Key_2', how='right')# comando how= por defecto 'inner'
```

##Join - Index Math

```python
left = pd.DataFrame({'A': ['A0','A1','A2'],
  'B':['B0','B1','B2']},
  index=['k0','k1','k2'])

right = pd.DataFrame({'C': ['C0','C1','C2'],
  'D':['D0','D1','D2']},
  index=['k0','k2','k3'])
```

```python
left.join(right, how='outer') #se pude cambiar el parametro how para hacer inner left right o outer Join
```
